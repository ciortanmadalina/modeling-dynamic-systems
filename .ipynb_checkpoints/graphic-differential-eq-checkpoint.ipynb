x(n+1) =  r * x (n) (1 - x(n))
0<= r <= 4
0 <= x (0) <= 1

r = [0, 1, 2, 2.5, 2.7, 3.5, 4] 

x0 = [0, 0.3, 0.5, 0.9, 1]

```python
rValues = [0, 1, 2, 2.5, 2.7, 3.5, 4]
x0Values = [0, 0.3, 0.5, 0.9, 1]
nbElements = 100
```

```python
def plotLogisticFunction(r, x0):
    x = [None] * nbElements
    x[0] = x0
    for i in range(1,nbElements):
        x[i] = r * x[i-1] * (1 - x[i-1])

    plt.figure(figsize=(12,8))
    plt.scatter(x = list(range(0, nbElements)), y = x, alpha=0.5, c= 'red', marker = '.')
    plt.title('r = ' + str(r) + ', x0 = ' + str(x[0]))
    plt.xlabel('position')
    plt.ylabel('value')
    plt.xticks(rotation=-45)
```

```python
for r in rValues:
    for x0 in x0Values:
        plotLogisticFunction(r, x0)
```
